# API Call

Call JOA government data API from the website below, using http requests python library in json data format. Since the data return a bunch of variables and we only need to select 'JobData' field of each job element.

Assuming clients are applicants like myself who holds a computer science degree, looking for a job associated with the degree. We found the seried number for computer science jobs is 1550 from JOA website. Details of the series number is here: http://www.opm.gov/policy-data-oversight/classification-qualifications/general-schedule-qualification-standards/#url=List-by-Occupational-Series

```python
import requests
import json

# set up for API Call

# base url: https://data.usajobs.gov/api/jobs
# HTTPS GET request
# request header content-type accepts application/json

request_token_url = 'https://data.usajobs.gov/api/jobs?Series=1550'
HEADERS = {'Content-Type': 'application/json'} 
```

```python
# get status of server response
res = requests.get(request_token_url)

status_object = json.loads(res.content);
# print(json.dumps(status_object, indent=4))
```

```python
# select only 'JobData' field in data 
job_data = status_object['JobData']
# print job_data
```

# Creating A DataFrame

From the selected field in the API call, we use pandas library to put the data nicely into a table and display the top 5 element of the table (for debugging). In a process of creating the table, we disregard columns that are irrelevant to our analysis and only include the ones we need for future analysis. We can also organize the columns such that we can see the job title first, organization name next, agency sub element after, and other information and so on.

```python
# create a data frame for json data 
from pandas import DataFrame, Series
import pandas as pd

frame = pd.DataFrame(job_data)

# frame columns with desired data
frame = frame[['JobTitle', 'OrganizationName', 'AgencySubElement', 'Locations', 'SalaryMax', 'SalaryMin', 'SalaryBasis', 'StartDate', 'EndDate']]
frame.head() # for debugging purposes, print the first 5 rows in the frame
```

# Find Least Popular Job

We assume that if a job is unpopular relatively than others, the job posting duration must be the longest. In other words, regardless of job posting start dates and end dats, if the duration of the start and the end date is the longest, we determine that the job is the least popular, having fewer applicants. We assume that the applicants interests of checking and applying for a job throughout the year is uniformly spread out.

Using datatime library, compute the number of days between a start and an end date and return the job tile of the longest date. 

```python
# find least popular jobs
from datetime import date
from datetime import datetime as dt
from operator import itemgetter

date_frame = frame[['StartDate', 'EndDate', 'JobTitle']]

def leastPopularJobs(data_frame):
    ''' iterate through each row in a data frame and calculate using datatime python library 
    to compute the number of days in between two dates. Add the number and the name of the job 
    in a dictionary. Sorted by the dictionary values, reversed in a decreasing order,
    adn return the key of the first element of the dictionary.'''
    
    start = data_frame['StartDate']
    end = data_frame['EndDate']
    job_title = data_frame['JobTitle']

    date_format = "%m/%d/%Y"
    total_duration = {}
    for i_start in start.index:
        start_date = dt.strptime(start[i_start], date_format)
        end_date = dt.strptime(end[i_start], date_format)
        delta = abs(end_date - start_date)
            
        # in order to prevent from adding the same job name to be considered as the same,
        # we add a counter 'i' at the end of job name if the key already exists in the dictionary
        i = 0
        job_name = job_title[i_start].lower()
        if (job_name in total_duration.keys()):
            total_duration[job_name + ' ' + str(i)] = delta
            i += 1
        else:
            total_duration[job_name] = delta
                
    return sorted(total_duration.items(), key=itemgetter(1))[::-1][0][0]

res = leastPopularJobs(date_frame)
print res
```

```python
# find the highest average salary
salary_frame = frame[['SalaryMax', 'SalaryMin', 'JobTitle']]

def min_max_average_salary(data_frame):
    '''calculate average salary for each job and return the names of the jobs with the highest and lowest average salary.'''
    
    min_salary = data_frame['SalaryMin']
    max_salary = data_frame['SalaryMax']
    job_title = data_frame['JobTitle']
    
    salary_table = {}
    for i in job_title.index:
        int_min = min_salary[i].split('.')[0][1:].replace(",", "")
        int_max = max_salary[i].split('.')[0][1:].replace(",", "")
        average = float((int(int_max) + int(int_min)/2))
        
        counter = 0
        job_name = job_title[i].lower()
        if (job_name in salary_table.keys()):
            salary_table[job_name + ' ' + str(counter)] = average
            counter += 1
        else:
            salary_table[job_name] = average
    print 'highest average: ' + sorted(salary_table.items(), key=itemgetter(1))[::-1][0][0]
    print 'lowest average: ' + sorted(salary_table.items(), key=itemgetter(1))[::-1][-1][0]
    
min_max_average_salary(salary_frame)
```
